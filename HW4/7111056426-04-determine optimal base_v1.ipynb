```python
import sys, os
import cv2
import numpy as np
import math
```

```python
if __name__ == '__main__':
    
    print("Input number of pixels in a cluster (n>=2) and the target notation (F>=4):")
    n, F = map(int, input().split())
    while(n<2 or F<4):
        if(n<2):
            print("pixels in a cluster (n) should >=2,please retry")
            print("Input number of pixels in a cluster (n>=2) and the target notation (F>=4):")
            n, F = map(int, input().split())
        elif(F<4):
            print("target notation (F>=4),please retry")
            print("Input number of pixels in a cluster (n>=2) and the target notation (F>=4):")
            n, F = map(int, input().split())
    print("The value of n & F are: ",n,F)
    
    b=math.ceil(F**(1/n))
    print("b",b)
    FBV=np.full(n,b)

    print("FBV",FBV)
    if n==0 or F==0:
        print("exit")
        exit()
    
```

```python
def cal_M(BV):
    M = 1
    for i in range(len(BV)):
        M*=BV[i]
    return M
FBV_M=cal_M(FBV)
print(FBV_M)
```

```python
def EMSE(OBV):
    emse_b = 0.0
    for b in range(len(OBV)):
        emse_b += ((OBV[b]**2)-((-2)**((OBV[b]+1)%2)))/12
    emse_obv = emse_b / len(OBV)
    return round(emse_obv,4)
```

```python
cur_CBV=np.copy(FBV)
cur_M=FBV_M
i_front=0
i_rear=len(cur_CBV)-1
stop=False
# record all M stop when M repeat
M_array=[]
# record M which M>=F 
CM_array=[]
# record EMSE which M>=F 
CEMSE_array=[]
# record CBV which M>=F
CBV_array=[]

while(stop==False):
    
    cur_M=cal_M(cur_CBV)
    if(np.any(M_array==cur_M)):
        # repeat
        print("repeat stop")
        stop=True
    else:
        if(cur_M>=F):
            CBV_array.append(np.array(cur_CBV))
            CEMSE_array.append(EMSE(cur_CBV))
            CM_array.append(cur_M)
        M_array.append(cur_M)
    
        if(cur_M>F):
            print("i_front",i_front)
            cur_CBV[i_front]-=1 
            if(i_front<len(cur_CBV)):i_front+=1
        elif(cur_M<F):
            cur_CBV[i_rear]+=1
            if(i_rear>0): i_rear-=1
        else:#temp_M==F
            print("M=F stop")
            stop=True
        

        
#     return cur_CBV,cur_M
print("cur_CBV=",cur_CBV,"M=",cur_M)
print("CEMSE_array",CEMSE_array)
print("CBV_array",CBV_array)
print("CM_array",CM_array)
best_emse=np.min(CEMSE_array)
print("best_emse",best_emse)
best_index=np.argmin(CEMSE_array)
print("best_index",best_index)
best_M=CM_array[best_index]
best_BV=CBV_array[best_index]
psnr=round(10 * np.log10(255 * 255 / best_emse),2)
diff=best_M-F

print("1.Optimal Base Vector OBV: ",best_BV)
print("2.Derived Notation M:",best_M)
print("3.Difference:",diff)
print("4.EMSE OBV:",best_emse)
print("5.PSNR",psnr)
```
